```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf
import statsmodels.api as sm 
```

```python
df=pd.read_csv('50_Startups.csv')
```

```python
df
```

```python
df.describe()
```

```python
df.columns
```

```python
plt.scatter(df['Marketing_Spend'],df['Profit'])
```

```python
ml1 = smf.ols('Profit~Marketing_Spend',data=df).fit() # regression model
```

```python
ml1.summary()
```

```python
plt.plot(df['Marketing_Spend'],0.2465*df['Marketing_Spend']+60000,c="orange")
plt.scatter(df['Marketing_Spend'],df['Profit'])
```

```python
sm.graphics.influence_plot(ml1)
```

```python
df2=df.drop([19,46])
```

```python
ml2 = smf.ols('Profit~Marketing_Spend',data=df2).fit()
ml2.summary()
```
